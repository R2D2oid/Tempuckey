```python
def run_cmd(cmd):
    process = subprocess.Popen(cmd, shell=True, stdout=subprocess.PIPE, stderr=subprocess.STDOUT)
    output, error = process.communicate()
    return output,error

def wait_folder_creation(path):
    import time
    while not os.path.exists(path):
        time.sleep(1)

def create_folder(path):
    if not os.path.exists(path):
        success = os.mkdir(path)
        wait_folder_creation(path)
        return success
```

## Rename video files to <small> video_FACEOFF_000num.mp4 </small> format

```python

def rename_videos(path):
    fnames = [f for f in listdir(path) if isfile(join(path,f)) and f.split('_')[0]!='video']
    for fname in fnames:
        new_fname = 'video_FACEOFF_{}'.format((fname.split('_')[0]).zfill(6))
        rename_file(path, fname, new_fname)
        
        
def rename_file (path, old_fname, new_fname):
    cmd = 'mv {}/{} {}/{}.mp4'.format(path, old_fname, path, new_fname)
    run_cmd(cmd)
    print(cmd)
    print()
```

```python
vids_path = '/usr/local/data01/zahra/datasets/Tempuckey/all_untrimmed_FACEOFF'
rename_videos(vids_path)
```
