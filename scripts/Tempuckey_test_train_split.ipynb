```python
import pandas as pd
```

```python
labels_path = '/usr/local/data02/zahra/datasets/Tempuckey/labels/ground_truth/tempuckey_ground_truth_labels.csv'
labels_test_path = '/usr/local/data02/zahra/datasets/Tempuckey/labels/ground_truth/tempuckey_ground_truth_labels_test.csv'
labels_train_path = '/usr/local/data02/zahra/datasets/Tempuckey/labels/ground_truth/tempuckey_ground_truth_labels_train.csv'

df = pd.read_csv(labels_path)
```

```python
### adding faceoff_ID
# print(df.columns)
# df.insert(0, 'faceoff_ID', range(0, len(df)))
# print(df.columns)
# df.to_csv(labels_path2, sep=',', index=False)
```

```python
### split test/train sets with 30/70 ratio
df_test = df.groupby('type').apply(lambda x: x.sample(frac=0.3, random_state=42))
df_test = df_test.reset_index(drop=True)

df_train = df[~df.faceoff_ID.isin(df_test.faceoff_ID)]
```

```python
df_test.to_csv(labels_test_path, sep=',', index=False)
df_train.to_csv(labels_train_path, sep=',', index=False)
```
